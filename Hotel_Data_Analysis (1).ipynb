## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Loading the Dataset

```python
df = pd.read_csv('hotel_booking.csv')
```

## Exploratory Data Analysis and Data Cleaning

```python
df.head(10)
```

```python
df.tail()
```

```python
df.shape
```

```python
# Since the columns 'name', 'email', 'phone-number' and 'credit_card' are of no particular use to us for this problem statement, I will remove them from the DataFrame.
df = df.drop(['name', 'email', 'phone-number', 'credit_card'], axis=1)
```

```python
df.columns
```

```python
df.info()
```

```python
# Since the 'reservation_status_date' column consists of object dtype instead of datetime dtype, therefore converting.
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df['reservation_status_date']
```

```python
# To understand the columns having non numerical data.
df.describe(include = 'object')
```

```python
# To figure out the unique data points in these non numerical columns.
for col in df.describe(include = 'object'):
    print(col)
    print(df[col].unique())
    print('*'*50)
```

```python
# Checking for missing values.
df.isnull().sum()
```

```python
# It can be seen that 'company' has too many missing values, therefore it is too difficult to handle. Also, 'agent' has a fair number of missing values and it isn't of much significance to our problem statement. Hence, dropping both these columns.
df = df.drop(['company', 'agent'], axis=1)

# Also removing the remaining rows with missing values in the 'children' and 'country' columns. removing about 500 datapoints from a total of 1 lac+ won't affent the quality of analysis as such.
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# To understand the columns having numerical data.
df.describe()
```

```python
# In the 'adr' column which represents average daily rate, the max is 5400 while the mean is a mere 102. Also, the 75th percentile is also just 126, making this max value fishy -
(df['adr'] > 1000).sum()
```

```python
# Hence, there is only 1 value of adr over 1000 i.e. the max of 5400, so it's most likely an outlier and hence is removed.
df = df[df['adr'] < 1000]
```

```python
# Alse rate can't be nagative, the min of -6.38 is also an outlier. Checking number of datapoints with negative date -
(df['adr'] < 0).sum()
```

```python
# Hence, there is only 1 value of adr below 0 i.e. the min of -6.38, so it's an outlier and hence is removed.
df = df[df['adr'] > 0]
df['adr'].describe()
```

## Data Analysis and Visualization

```python
# Checking what % of reservations was cancelled. In the 'is_canceled' column, 1 represents a canceled reservation, 0 the opposite.
can_perc = df['is_canceled'].value_counts(normalize = True)
can_perc
# Hence it can be seen that about 37% of reservations were cancelled, which is a huge number when taken into perspective. So, this issue is a critical one, that needs to be addressed.
```

```python
# Visualizing the 'is_canceled' column.
plt.figure(figsize=(5,4))
plt.bar(x = ['Not Canceled', 'Canceled'],
        height = df['is_canceled'].value_counts(),
        edgecolor = 'k',
        width = 0.5)
plt.xlabel('Status of Booking', fontweight = 'bold')
plt.ylabel('Number of Bookings', fontweight = 'bold')
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.title('Reservation Status Count', fontweight = 'bold', fontsize = 14)
plt.show()
```

```python
# Analyzing the distrubution of canceled and non-canceled bookings in hotels as compared to those in resorts.
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
labels = ['Not Canceled', 'Canceled']
plt.legend(labels)
plt.xlabel('Hotel', fontweight = 'bold')
plt.ylabel('Number of Bookings', fontweight = 'bold')
plt.title('Reservation Status Count in Hotels vs Resorts', fontweight = 'bold', fontsize = 14)
```
