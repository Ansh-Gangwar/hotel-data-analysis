## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Loading the Dataset

```python
df = pd.read_csv('hotel_booking.csv')
```

## Exploratory Data Analysis and Data Cleaning

```python
df.head(10)
```

```python
df.tail()
```

```python
df.shape
```

```python
# Since the columns 'name', 'email', 'phone-number' and 'credit_card' are of no particular use to us for this problem statement, I will remove them from the DataFrame.
df = df.drop(['name', 'email', 'phone-number', 'credit_card'], axis=1)
```

```python
df.columns
```

```python
df.info()
```

```python
# Since the 'reservation_status_date' column consists of object dtype instead of datetime dtype, therefore converting.
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df['reservation_status_date']
```

```python
# To understand the columns having non numerical data.
df.describe(include = 'object')
```

```python
# To figure out the unique data points in these non numerical columns.
for col in df.describe(include = 'object'):
    print(col)
    print(df[col].unique())
    print('*'*50)
```

```python
# Checking for missing values.
df.isnull().sum()
```

```python
# It can be seen that 'company' has too many missing values, therefore it is too difficult to handle. Also, 'agent' has a fair number of missing values and it isn't of much significance to our problem statement. Hence, dropping both these columns.
df = df.drop(['company', 'agent'], axis=1)

# Also removing the remaining rows with missing values in the 'children' and 'country' columns. removing about 500 datapoints from a total of 1 lac+ won't affent the quality of analysis as such.
df = df.dropna()
```

```python
df.isnull().sum()
# So no more missing vallues left
```

```python
# To understand the columns having numerical data.
df.describe()
```

```python
# In the 'adr' column which represents average daily rate, the max is 5400 while the mean is a mere 102. Also, the 75th percentile is also just 126, making this max value fishy -
(df['adr'] > 1000).sum()
```

```python
# Hence, there is only 1 value of adr over 1000 i.e. the max of 5400, so it's most likely an outlier and hence is removed.
df = df[df['adr'] < 1000]
```

```python
# Alse rate can't be nagative, the min of -6.38 is also an outlier. Checking number of datapoints with negative date -
(df['adr'] < 0).sum()
```

```python
# Hence, there is only 1 value of adr below 0 i.e. the min of -6.38, so it's an outlier and hence is removed.
df = df[df['adr'] > 0]
df['adr'].describe()
```

## Data Analysis and Visualization

```python
# Checking what % of reservations was cancelled. In the 'is_canceled' column, 1 represents a canceled reservation, 0 the opposite.
can_perc = df['is_canceled'].value_counts(normalize = True)
can_perc
# Hence it can be seen that about 37% of reservations were cancelled, which is a huge number when taken into perspective. So, this issue is a critical one, that needs to be addressed.
```

```python
# Visualizing the 'is_canceled' column.
plt.figure(figsize=(5,4))
plt.bar(x = ['Not Canceled', 'Canceled'],
        height = df['is_canceled'].value_counts(),
        edgecolor = 'k',
        width = 0.5)
plt.xlabel('Status of Booking', fontweight = 'bold')
plt.ylabel('Number of Bookings', fontweight = 'bold')
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.title('Reservation Status Count', fontweight = 'bold', fontsize = 14)
plt.show()
```

```python
# Analyzing the distrubution of canceled and non-canceled bookings in hotels as compared to those in resorts.
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues', edgecolor = 'k')
labels = ['Not Canceled', 'Canceled']
plt.legend(labels)
plt.xlabel('Hotel', fontweight = 'bold')
plt.ylabel('Number of Bookings', fontweight = 'bold')
plt.title('Reservation Status Count in Hotels vs Resorts', fontweight = 'bold', fontsize = 14)
plt.show()
```

```python
# Calculating percentage of bookings that were cancelled in city hotels.
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
# 42% of city hotel bookings were cancelled.
```

```python
# Calculating percentage of bookings that were cancelled in resort hotels.
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
# 28% of resort hotel bookings were cancelled.
```

```python
# Analyzing the factors affecting cancellations in city and resort hotels. First comparing the nightly rates.
# Grouping by 'reservation_status_date' since there are multiple data entries for a single date, so we need mean nightly rate for each night.
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Visualizing these average daily rates
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotels', fontsize = 21, fontweight = 'bold')
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel', color = 'red')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel', color = 'blue')
plt.legend(fontsize = 15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Date', fontsize = 17, fontweight = 'bold')
plt.ylabel('Rate', fontsize = 17, fontweight = 'bold')

plt.show()

# Hypothesis is that the spikes in the rate in both resort and city hotels are on weekends or other public holidays.
# It can be seen that the average daily rate for resort hotels undergoes a greater fluctuation as compared to city hotels.
```

```python
# Finding which months have greatest number of reservations and cancellations
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright', edgecolor = 'k')
labels = ['Not Canceled', 'Canceled']
plt.legend(labels)
plt.xlabel('Month', fontsize = 14, fontweight = 'bold')
plt.ylabel('Number of Reservations', fontsize = 14, fontweight = 'bold')

plt.title('Reservation Status Count for Different Months', fontweight = 'bold', fontsize = 18)
plt.show()

# It can be seen that the greatest number of cancellations have been made in January, reservations in August.
# It can be seen that the lowest number of cancellations have been made in September, reservations in December.
# Peculiar that Jan and Dec have very low reservations as well as high number of cancellations.
# Also, Aug and Sept have very high reservations as well as low number of cancellations.
```

```python
# The reason for the above mentioned peculiarity could be low prices in months of Aug and Sept, high prices in Jan and Dec.
# Low prices motivate the customers to go ahead with the reservation and not cancel, vice versa is valid too.
# Therefore, plotting average daily rate for each month, for cancelled bookings

plt. figure(figsize = (16,8))
sns.barplot(x = 'month', y = 'adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].mean().reset_index(), edgecolor = 'k')
plt.xlabel('Month', fontsize = 16, fontweight = 'bold')
plt.ylabel('Average Daily Rate', fontsize = 16, fontweight = 'bold')

plt.title('Average Daily Rate per Month for Cancelled Bookings', fontweight = 'bold', fontsize = 20)
plt.show()

# Hence it can be seen that Aug and Sept have comparatively lower adr on cancelled bookings, therefore the overall cancellations in these months is lower.
# Also adr is highest in Jan for cancelled bookings, therefore the highest cancellations are in Jan.
```

```python
# Checking country wise statistics of cancellations. Looking at top 10 out of the 177 countries in the dataset.
cancelled_data = df[df['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[0:10]
top_10_country
```

```python
# Visualizing this data
plt.figure(figsize = (8,8))
plt.title('Top 10 Countries with Most Cancelled Reservations', fontweight = 'bold', fontsize = 14)
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
# Here it can be seen that Portugal is the country with most number of cancellations, a staggering 70%.
# Great Britain, Spain and France also have a substantial % of cancelled bookings, about 5%.
```

```python
# Checking from which mode majority customers come (Online Agents / Offline Agents/ Groups/ etc).
df['market_segment'].value_counts(normalize = True)
# Can be seen that about 48% of bookings are made through the Online Mode, it is the most preferred.
# Could be due to convenience, promotional offers and better prices offered for online bookings
```

```python
# Checking from which mode majority of the cancelled bookings were made.
cancelled_data['market_segment'].value_counts(normalize=True)
# Therefore about 47% of the cancelled bookings were made online. 
# It is possible that online websites portray an unfair and improved image of hotels, upon actually visiting the hotel customers feel disappointed.
```

```python
# Comparing adr of cancelled and non-cancelled bookings.
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate for Cancelled vs Not Cancelled', fontsize = 20, fontweight = 'bold')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'Not Cancelled', color = 'blue')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'Cancelled', color = 'red')
labels = ['Not Canceled', 'Canceled']
plt.legend(labels, fontsize = 14)
plt.xlabel('Date', fontsize = 15, fontweight = 'bold')
plt.ylabel('Number of Reservations', fontsize = 15, fontweight = 'bold')
plt.show()
```

```python
# As seen in above plot, data is irregular before Sept 2015 and after Sept 2017. So, considering data only in this timeframe.
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2015-09-01') & (cancelled_df_adr['reservation_status_date'] < '2017-09-30')]
cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2015-09-01') & (not_cancelled_df_adr['reservation_status_date'] < '2017-09-30')]
not_cancelled_df_adr.reset_index(inplace=True)
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate for Cancelled vs Not Cancelled (Sept 2015 - Sept 2017)', fontsize = 20, fontweight = 'bold')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'Not Cancelled', color = 'blue')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'Cancelled', color = 'red')
labels = ['Not Canceled', 'Canceled']
plt.legend(labels, fontsize = 14)
plt.xlabel('Date', fontsize = 15, fontweight = 'bold')
plt.ylabel('Number of Reservations', fontsize = 15, fontweight = 'bold')
plt.show()
# It can be seen that for most of the period, adr of cancelled bookings was greater than that of non-cancelled ones.
```

```python
# Checking average no. of days in waiting list for canceled and non canceled bookings.
df_wait = df.groupby('is_canceled')[['days_in_waiting_list']].mean()
df_wait
# Avg waiting days is slighly greater than twice for canceled bookings as compared to non canceled
```

```python
# Visualizing this data
plt.figure(figsize = (6,4))
plt.bar(x = ['Not Canceled', 'Canceled'], height = df_wait['days_in_waiting_list'], edgecolor = 'k', width = 0.4)
plt.xlabel('Booking Status', fontsize = 12, fontweight = 'bold')
plt.ylabel('Days in Waiting List', fontsize = 12, fontweight = 'bold')

plt.title('Average No. of Days in Waiting List for Canceled vs Non Canceled', fontweight = 'bold', fontsize = 15)
plt.show()

# Hence, higher waiting times significantly affect bookings and increase the number of cancellations.
```
